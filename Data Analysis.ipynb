```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing  import LabelEncoder
import matplotlib.pyplot as plt
df = pd.read_csv('insurance.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
miss = (df['age'].isnull().sum())/len(df)*100
print("Percentage of missing values at age = ", miss)
```

```python
miss = (df['smoker'].isnull().sum())/len(df)*100
print("Percentage of missing values at smoke = ", miss)
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
miss = (df['age'].isnull().sum())/len(df)*100
print("Percentage of missing values at age = ", miss)
```

```python
df['smoker'] = df['smoker'].fillna(df['smoker'].mode()[0])
miss = (df['smoker'].isnull().sum())/len(df)*100
print("Percentage of missing values at smoker = ", miss)
```

```python
df['sex'] = LabelEncoder().fit_transform(df['sex'])
df['smoker'] = LabelEncoder().fit_transform(df['smoker'])
print(df)
```

```python

plt.hist(df['bmi'])
plt.title('BMI Distribution Histogram')
plt.xlabel('bmi')
plt.ylabel('Frequency')
plt.show()

mx = df['bmi'].max()
mn = df['bmi'].min()
df['bmi'] = (df['bmi'] - mn) / (mx - mn)


plt.hist(df['bmi'])
plt.title('BMI Distribution Histogram')
plt.xlabel('bmi')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['children'])
plt.title('children Distribution Histogram')
plt.xlabel('children')
plt.ylabel('Frequency')
plt.show()

df['children'] = np.log1p(df['children'])

plt.hist(df['children'])
plt.title('children Distribution Histogram')
plt.xlabel('children')
plt.ylabel('Frequency')
plt.show()


```

```python
plt.hist(df['charges'])
plt.title('charges Distribution Histogram')
plt.xlabel('charges')
plt.ylabel('Frequency')
plt.show()

df['charges'] = np.log1p(df['charges'])

plt.hist(df['charges'])
plt.title('charges Distribution Histogram')
plt.xlabel('charges')
plt.ylabel('Frequency')
plt.show()

```
